```python
#Q1
import requests
from bs4 import BeautifulSoup
import pandas as pd

books = []

url = "https://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

for book in soup.find_all('article', class_='product_pod'):
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text.strip()
    availability = book.find('p', class_='instock availability').text.strip()
    star_rating = book.find('p', class_='star-rating')['class'][1]
    
    books.append({
        'Title': title,
        'Price': price,
        'Availability': availability,
        'Star Rating': star_rating
    })

for page in range(2, 51):
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    for book in soup.find_all('article', class_='product_pod'):
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text.strip()
        availability = book.find('p', class_='instock availability').text.strip()
        star_rating = book.find('p', class_='star-rating')['class'][1]
        
        books.append({
            'Title': title,
            'Price': price,
            'Availability': availability,
            'Star Rating': star_rating
        })

df = pd.DataFrame(books)
df.to_csv('books.csv', index=False)
print("books.csv file created successfully.")
```

```python
#Q2
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.imdb.com/chart/top/")
time.sleep(3)

movies = []
rows = driver.find_elements(By.XPATH, '//li[@class="ipc-metadata-list-summary-item"]')

for rank, row in enumerate(rows, start=1):
    try:
        title = row.find_element(By.XPATH, './/h3').text.split('. ')[1]
        year = row.find_element(By.XPATH, './/span[contains(@class,"title-metadata-item")]').text
        rating = row.find_element(By.XPATH, './/span[contains(@class,"ipc-rating-star--rating")]').text
        movies.append([rank, title, year, rating])
    except:
        continue

driver.quit()

df = pd.DataFrame(movies, columns=["Rank", "Movie Title", "Year", "IMDB Rating"])
df.to_csv("imdb_top250.csv", index=False)
print("imdb_top250.csv file created successfully.")
```

```python
#Q3
import requests
from bs4 import BeautifulSoup
import pandas as pd

r = requests.get("https://www.timeanddate.com/weather/")
soup = BeautifulSoup(r.content, 'html.parser')

table = soup.find('table', {'class': 'zebra fw tb-theme'})
weather = []
if table:
    rows = table.find_all('tr')  
    for row in rows:
        cols = row.find_all('td')
        if len(cols) >= 3:  
            city = cols[0].text.strip()
            temp = cols[3].text.strip()
            
            img = cols[2].find("img")
            if img and img.has_attr('alt'):
                desc = img['alt'].strip()
                if "clear" in desc or "sunny" in desc: condition = "Clear"
                elif "rain" in desc or "drizzle" in desc: condition = "Rainy"
                elif "cloud" in desc or "overcast" in desc: condition = "Cloudy"
                elif "snow" in desc or "sleet" in desc: condition = "Snowy"
                elif "storm" in desc or "thunder" in desc: condition = "Stormy"
                elif "fog" in desc or "mist" in desc: condition = "Foggy"
                else: condition = "Other"
            else:
                condition = "Other"
            weather.append([city, temp, condition])

df_weather = pd.DataFrame(weather, columns=['City', 'Temperature', 'Condition'])
df_weather.to_csv('weather.csv', index=False)
print("weather.csv file created successfully.")
```
